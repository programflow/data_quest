# Guided Project: Visualizing Earnings Based On College Majors

## 1. Introduction

In this guided project, I'm going to explore the pandas plotting functionality with the Jupyter Notebook interface and how they allows us to visualize that data quickly in order to explore the data more easily.

The dataset I will work with is the job outcomes of students who graduated from college between 2010 and 2012. American Community Survey conducted the survey and agregated the data, and released the original data on job outcomes. FiveThirtyEight cleaned the dataset and released it on their Github repo.

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more.

The columns in the dataset:
* Rank - Rank by median earnings (the dataset is ordered by this column)
* Major_code - Major code
* Major - Major description
* Major_category - Category of major
* Total - Total number of people with major.
* Sample_size - Sample size (unweighted) of full-time
* Men - Male graduates
* Female - Female graduates
* ShareWomen - Women as share of total
* Employed - Number employed
* Median - Median salary of full-time, year-round workers
* Low_wage_jobs - Number in low-wage service jobs.
* Full-time - Number employed 35 hours or more
* Part_time - Number employed less than 35 hours

Using Visualizations we can explore different questions from the dataset like:
* Do students in more popular majors make more money?
    * Using scatter plots
* How many majors are predominantly male? Predominantly female?
    * using histograms
* Which category of majors have the most students?
    * Using bar plots

```python
# We're going to start by importing the libraries we need 
# and remove rows containing null values

# import libraries
import pandas as pd
import matplotlib.pyplot as plt
# Run Jupyter magic so plots are displayed inline
%matplotlib inline
```

```python
# Now we want to read in the dataset into a DataFrame
recent_grads = pd.read_csv('recent-grads.csv')
```

```python
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
recent_grads.describe()
```

```python
# Now we are going to drop rows with missing vlaues because 
# Matplotlib expects columns to having matching lengths and missing
# values will cause maplotlib to throw errors

raw_data_count = len(recent_grads)
recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads)
print(raw_data_count)
print(cleaned_data_count)
```

```python
# Only one row cotained missing values and was dropped
```

## 2. Pandas, Scatter Plots

The plotting functionality is contained with DataFrame.plot(). We use the 'kind' parameter to specify the type of plot we want. We use 'x' and 'y' to specify the data we want on each axis.

```python
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter')
```

```python
# There are a few extra parameters that we can use to tweak our plot
# title and figsize

recent_grads.plot(x='Sample_size', y='Employed', kind='scatter', title='Employed vs. Sample_size', figsize=(5,10))
```

```python
# we can access the underlying matplotlib Axes object by assigning the return value to a variable:
ax = recent_grads.plot(x='Sample_size', y='Employed', kind='scatter')
ax.set_title('Employed vs. Sample_size')
```

```python
# we are going to explore some different relationships
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Median vs. Sample_size')
```

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title('Unemployment_rate vs. Sample_size')
```

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Median vs. Full_time')
```

```python
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax.set_title('Unemployment_rate vs. ShareWomen')
```

```python
ax = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax.set_title('Median vs. Men')
```

```python
ax = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax.set_title('Median vs. Women')
```

Now we want to try answer some questions with the plots.
* Do students in more popular majors make more money?
    * The more popular majors don't seem to make more money. It seems as if the less popular majors make more money which would make sense with regards to scarcity.
* Do students that majored in subjects that were majority female make more money?
    * It seems like it doesn't seem to matter all that much
* Is there any link between the number of full-time employees and median salary?
    * The higher number of full-time employees the less the median salary is it seems.
 

## 3. Pandas, Histograms

```python
# if we want to explore the distribution of values within a single
# column we can use Series.plot() and set kind=hist

recent_grads['Sample_size'].plot(kind='hist')
```

```python
# instead we can use Series.hist() which allows use to control the binning
recent_grads['Sample_size'].hist(bins=25, range=(0, 5000))
```

```python
ax = recent_grads['Sample_size'].hist(bins=25, range=(0, 5000))
ax.set_title('Sample size')
```

```python
ax = recent_grads['Median'].hist(bins=25, range=(0, 110000))
ax.set_title('Median Salary')
```

```python
ax = recent_grads['Employed'].hist(bins=25, range=(0, 300000))
ax.set_title('Employed')
```

```python
ax = recent_grads['Full_time'].hist(bins=25, range=(0, 250000))
ax.set_title('Employed Full-time')
```

```python
ax = recent_grads['ShareWomen'].hist(bins=25, range=(0, 1))
ax.set_title('ShareWomen')
```

```python
ax = recent_grads['Unemployment_rate'].hist(bins=25, range=(0, .2))
ax.set_title('Unemployment rate')
```

```python
ax = recent_grads['Men'].hist(bins=25, range=(0, 175000))
ax.set_title('Men in Major')
```

```python
ax = recent_grads['Women'].hist(bins=25, range=(0, 300000))
ax.set_title('Women in major')
```

## 4. Pandas, Scatter Matrix Plot

A scatter matrix plot combines both scatter plots histograms into one grid of plots and allows use to explore potential relationships and distributions simultaneously. It consists of n by n plots. The diagnol plots are histograms and the non-diagnol are scatter.

```python
from pandas.plotting import scatter_matrix
```

```python
# we can use scatter_matrix()
scatter_matrix(recent_grads[['Women', 'Men']], figsize=(10, 10))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10, 10))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10, 10))
```
