# 2. Cleaning Column Names

Now we want to convert the column names from camelcase to snakecase and reword some of the columns names based on the data dictionary to be more descriptive.

```python
col_names = autos.columns
new_names = ["date_crawled","name","seller","offer_type","price",
            "abtest", "vehicle_type", "registration_year", "gearbox",
            "power_ps","model", "odometer", "registration_month","fuel_type",
            "brand", "unrepaired_damage", "ad_created", "nr_of_pictures",
            "postal_code", "last_seen"]
col_map = {}
for i in range(0,len(col_names)):
    col_map[col_names[i]] = new_names[i]
    
```

```python
autos.columns = col_names.map(col_map)
autos.columns
```

```python
autos.head()
```

to make it easier to read and use we changed the column names to snakecase and reduced the words that are contain in certain column names for brevity.

# 3. Initial Exploration and Cleaning

Now we will do some more data exploration to determine what other cleaning tasks need to be done.

Things to check for:
1. Text columns where almost all thevalues are the same. if there is little or no variability the information isn't particularly useful.
2. Examples of numeric data stored as text which can be cleaned and converted


```python
autos.describe()
```

nr_of_pictures column seems to have only 1 value

registration_month happens to have a 0 as a min which means that at least one of the cars has 0 as a value for that column. That doesn't really correspond to any month so we need to deal with that.

registration_year happens to have a min of 1000 and a max of 9999 whic don't really makes sense because cars have only been around for less than 200 years and we are only in year 2020. so registering a car in the future doesn't make sense. We're gonna have to establish a time period for what range or registration years are valid.

The odometer and price have some text mixed with numeric values so we're going to have clean those columns so we'll start of by cleaning those since they're the simplest.

```python
autos["price"] = autos["price"].str.replace("$","")
autos["price"] = autos["price"].str.replace(",","")
autos["price"] = autos["price"].astype(int)
autos["odometer"] = autos["odometer"].str.replace("km","")
autos["odometer"] = autos["odometer"].str.replace(",","")
autos["odometer"] = autos["odometer"].astype(int)
autos = autos.rename(columns={"price":"price_dollars",
                      "odometer":"odometer_km"
                     })
```

Because these are listings for 2016 the we can choose that the only valid registration years are between 1900-2016.

We need to test this first so we'll include them and then exclude them to see if it makes a difference.


```python
autos["registration_year"].value_counts(normalize=True).sort_index(ascending=True)
```

```python
reg_year_bool = (autos["registration_year"] >=1900) &(autos["registration_year"]<=2016)
```

# 7. Exploring Price by Brand

When working with data on cars, it's natural to explore variations across different car brands. We can use aggregation to understand the brand column.

```python
autos["brand"].unique()
```

```python
autos["brand"].value_counts(normalize=True).sort_values(ascending=False)
```

We'll aggregate in two ways first by the top 20 and then by any brand contributing 5% or more.

```python
top_20_brands = autos["brand"].value_counts().sort_values(ascending=False).head(20).index
```

```python
aggregate_data = {}
for b in top_20_brands: 
    selected_rows = autos[autos["brand"] == b]
    mean_price = selected_rows["price_dollars"].mean() 
    aggregate_data[b] = mean_price
```

```python
aggregate_data
```

```python
brands_5_percent = autos["brand"].value_counts(normalize=True).sort_values(ascending=False).head(6).index
```

```python
aggregate_data_2 = {}
for b in brands_5_percent: 
    selected_rows = autos[autos["brand"] == b]
    mean_price = selected_rows["price_dollars"].mean() 
    aggregate_data_2[b] = mean_price
```

```python
aggregate_data_2
```

Comparing the 2 ways we aggregated the data based on brand we can see that on average the mean price is a bit higher for both the top brands. one of the brands not contributing at least top 5% of sales was citreon sonstige_autos and mini which had means that were quite high. Higher than the rest of the brands.

# 8. Storing aggregate data in a dataframe

We are now going take aggregate data and convert them to data frames. I'll stick with the top 20 aggregate data. We also want to get the mean mileage of all the top 20 brands and make them into a single data frame.

```python
mileage_data ={}
for b in top_20_brands: 
    selected_rows = autos[autos["brand"] == b]
    mean_mileage = selected_rows["odometer_km"].mean() * .6213
    mileage_data[b] = mean_mileage
```

```python
mileage_series = pd.Series(mileage_data)
```

```python
price_series = pd.Series(aggregate_data)
```

```python
brands_mean_df = pd.DataFrame(mileage_series, columns=["mean_mileage"])
```

```python
brands_mean_df["mean_price"] = price_series
```

```python
brands_mean_df
```

It seems that on average regardles of mileage the closer the cars are to the top the more expnsive they'll be.
