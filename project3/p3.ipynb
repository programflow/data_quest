# 2. Cleaning Column Names

Some more observations we can make are that the atatset contains 20 columns, 15 of those are strings. Some columns have null values, but none have more than ~20% null values. The column name follow the camelcase format instead of snakecase which means we can't just replace spaces with underscores.

We are going to convert the column names from camelcase to snakecase and reword some of the column names based on the data dictionary to be more descriptive.

```python
autos.columns

new_columns = ['name', 'seller', 'offer_type', 'price', 'ab_test', 
              'vehicle_type', 'registration_year', 'gearbox', 'power_ps',
              'model', 'odometer', 'registration_month', 'fuel_type', 'brand',
              'unrepaired_damage', 'ad_created', 'nr_of_pictures', 'postal_code',
              'last_seen'
             ]

autos.columns = new_columns

autos.head
```

# 3. Initial Exploration and Cleaning

We'll explore the data a bit more to see what other cleaning tasks need to be done. So we'll start by looking for tesxt columns where all or almost all values are the same. These can often be dropped as they don't have sueful information for analysis. We'll look for examples of numeric data stored as text which can be cleaned and converted.




```python
autos.describe(include='all')
```

Notes:

-> The columsn with too few values are seller, offer type, ab_test, gearbox, unrepaired damage

-> The columns that need more investigation are vehicle_type, registration_year, power_ps, odometer, registration month, fue_type, brand, nr_of_pictures, and postal_code.

-> The columns that have are stored as text that need to be cleaned are price, and odometer.

-> The columns that I'm a bit uncertain about are registration_year, power_ps, registration_month, nr_of_pictures, postal_code, and unrepaired_damage.

So the price and odometer columns can be cleaned relatively easily. So we'll take of those type by removing any non-numeric characters, convert the column to a numeric dtype, and rename them appropriately.

```python
autos['price'] = autos['price'].str.replace(",","")
autos['price'] = autos['price'].str.replace("$","")
autos['price'] = autos['price'].astype(int)
autos['odometer'] = autos['odometer'].str.replace(",","")
autos['odometer'] = autos['odometer'].str.replace("km","")
autos['odometer'] = autos['odometer'].astype(int)
```

```python
autos.rename({'odometer':'odometer_km'}, axis = 1, inplace = True)
```
